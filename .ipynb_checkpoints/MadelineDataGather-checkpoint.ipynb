```python
import os
import pandas as pd
import csv

csvpath = "../parking-citations.csv"

with open(csvpath, newline='') as csvfile:

    # CSV reader specifies delimiter and variable that holds contents
    csvreader = csv.reader(csvfile, delimiter=',')

    print(csvreader)

    # Read the header row first (skip this step if there is now header)
    csv_header = next(csvreader)
    print(f"CSV Header: {csv_header}")
```

```python
#droppping any row that has a null value gives us no rows

clean_vio_df = violations_df.dropna(how="any")
clean_vio_df.count()
```

```python
grouped_vio_df = violations_df.groupby(['Violation Description']).count()
grouped_vio_df.head(10)
```

```python
#clean time

violations_df['Date_Clean'] =violations_df['Issue Date'].str.slice(0, 10)
#violations_df.head()

#prepping set for violations by percentages

vio_type_df = violations_df[['Date_Clean', 'Violation code', 'Violation Description', 'Fine amount']].sort_values(by=['Date_Clean'])
vio_type_df.head()
                      
```
